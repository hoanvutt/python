```python
#Importing Libraries

import numpy as np

import pandas as pd

from os import path

from PIL import Image

import matplotlib.pyplot as plt

%matplotlib inline

from wordcloud import WordCloud
```

```python
#Importing Dataset
# df = pd.read_csv("worldcloud.csv")
df = pd.read_csv("worldcloud_full.csv")
```

```python
#Checking the Data
df.head(10)
```

```python
#Creating the text variable

text2 = " ".join(user_name for user_name in df.user_name)
```

```python
print(text2)
```

```python
# Creating word_cloud with text as argument in .generate() method

word_cloud2 = WordCloud(collocations = False, background_color = 'white').generate(text2)
```

```python
# Display the generated Word Cloud

plt.imshow(word_cloud2, interpolation='bilinear')

plt.axis("off")

plt.show()
```

```python
mask = np.array(Image.open("house_icon.png"))
wc = WordCloud(background_color='white', width = 1000, height=1000, margin=2, mask=mask)
```

```python
# print(df.to_dict())
df_data = df.set_index('user_name').to_dict()
df_data = df_data["page_rank"]
print(df_data)
```

```python
wc.fit_words(df_data)
# wc.to_file('wc.png')
```

```python
#  plt.figure()
plt.figure(figsize=[15,15])
plt.imshow(wc, interpolation='bilinear')
plt.axis("off")
plt.show()
# wc.to_file('wc.png')
wc.to_file('wc_full.png')
```
